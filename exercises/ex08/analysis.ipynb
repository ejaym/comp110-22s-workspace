We then must read the data found in `survey.csv` in order to later process it into a table.

Now that the data is read, it is in a row based configuration rather than a column based configuration. This next cell will convert it from a row to a column based so that the data is easier to work with.

```python
data_cols: dict[str, list[str]] = columnar(data_rows)
if len(data_cols.keys()) == 0:
    print("Complete your implementation of columnar in data_utils.py")
    print("Be sure to follow the guidelines above and save your work before re-evaluating!")
else:
    print(f"{len(data_cols.keys())} columns")
    print(f"{len(data_cols['row'])} rows")
    print(f"Columns names: {data_cols.keys()}")
```

Now that the data is in a column based form we can now select the columns of information that we specifically want to see using the `select` function, in this case it will be the year column. We will then put this into tabulate so it can create a nice looking table out of the data we feed into it.

```python
selected_data: dict[str, list[str]] = select(data_cols, ["year"])

tabulate(selected_data, selected_data.keys(), "html")
```

While the table above looks overwhelming at first, by using the `count` function, we can simply count the number of times each number appears within the table above, saving us the work of having to count each graduation year individually.

```python
year_counts: dict[str, int] = count(selected_data["year"])
print(f"Year_counts: {year_counts}")
```

In order to better visualize this data, we will compile these numbers into a bar graph using the `matplotlib` third party function.

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
fig.suptitle("Amount of students in COMP 110 by Graduation Year")


axes[0].set_title("By Class")
axes[0].bar(year_counts.keys(), year_counts.values())
axes[0].tick_params(axis='x', labelrotation = 45)


```

As a final step, the following function (`upper_lower`) will seperate the graduating years into 2 categories: those graduating in 2022 or 2023 - the upperclassmen, and those graduating in years beyond 2023 - the lowerclassmen. This allows the reader to understand the full picture of what is truly the issue in regards to classroom numbers in COMP110.

```python
upper_lower(selected_data)
```
